```python
import numpy as np
import matplotlib.pyplot as plt

# Population parameters
population_size = 1000
true_proportion_smoking = 0.395

# Generate population data
population_data = np.random.choice([0, 1], size=population_size, p=[1-true_proportion_smoking, true_proportion_smoking])

# Number of samples and sample size
num_samples = 1000
sample_size = 100

# Generate random samples and calculate sample proportions
sample_proportions = []
for _ in range(num_samples):
    sample = np.random.choice(population_data, size=sample_size, replace=False)
    sample_proportion = np.sum(sample) / sample_size
    sample_proportions.append(sample_proportion)

# Plot the distribution of sample proportions
plt.hist(sample_proportions, bins=30, density=True, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Sample Proportion')
plt.ylabel('Density')
plt.title('Sampling Distribution of Sample Proportions')
plt.grid(True)
plt.show()

```
