```python
population=np.random.uniform(0,100,1000)
print("Populations ")
plt.hist(population)
plt.show()
mean=np.mean(population)
variance=np.var(population)
samples,means,variance=[],[],[]
for i in range(100):
  s=random.sample(tuple(population),20)
  samples.append(s)
  means.append(np.mean(s))
  variance.append(np.var(s))
print("Sample Means")
plt.hist(means)
plt.show()
print(f"Mean of samples = {np.mean(means)}")
print(f"Mean = {np.mean(population)}")
```
