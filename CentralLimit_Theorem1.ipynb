Q1 - The lifetime of a certain brand of electric bulb may be considered a random variable with mean 1200 hrs and std deviation 250 hrs. Using the central limit theorem, find the probability that the average lifetime of 60 bulbs exceeds 1250 hrs.

 Q2- If x1, x2,….xn are Poisson variates with the parameter lambda =2, use central limit theorem to estimate P(120<Sn<160), where Sn= x1 + x2 + …….+ xn and n=75

```python
import numpy as np
import math
from scipy.stats import norm
lambda_ = 2
n = 75
mean = n * lambda_
std_dev = math.sqrt(n * lambda_)
p = norm.cdf((160- mean)/std_dev) - norm.cdf((120- mean)/ std_dev)
print(p)
```
