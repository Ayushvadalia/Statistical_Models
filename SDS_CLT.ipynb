```python
# 20

import random
import matplotlib.pyplot as plt

# Create a population of 100 numbers
population = list(range(1, 101))

# Get the sample size and number of samples from the user
n = int(input("Enter the sample size: "))
num_samples = int(input("Enter the number of samples: "))

# Create a list to store the sample means
sample_means = []

# Draw num_samples samples of size n from the population and calculate the mean of each sample
for i in range(num_samples):
  sample = random.sample(population, n)
  sample_mean = sum(sample) / len(sample)
  sample_means.append(sample_mean)

# Plot the distribution of the sample means
plt.hist(sample_means)
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

The code first generates a population of 100 numbers. Then, the user is prompted to enter the sample size and the number of samples. As the number of samples increases, the distribution of the sample means becomes more normal, even though the population is not normally distributed. This is consistent with the Central Limit Theorem, which states that the distribution of sample means will be approximately normal, regardless of the distribution of the population, as long as the sample size is sufficiently large.


```python
# 23

import random
import matplotlib.pyplot as plt
# Set the population size and true population proportion
population_size = 1000
true_population_proportion = 0.395
# Create a population of smokers and non-smokers
population = [True] * int(population_size * true_population_proportion) + [False] * int(population_size * (1 - true_population_proportion))
# Get the sample size and number of samples from the user
sample_size = int(input("Enter the sample size: "))
num_samples = int(input("Enter the number of samples: "))
# Create a list to store the sample proportions
sample_proportions = []

# Draw num_samples samples of size n from the population and calculate the proportion of smokers in each sample
for i in range(num_samples):
  sample = random.sample(population, sample_size)
  sample_proportion = sum(sample) / len(sample)
  sample_proportions.append(sample_proportion)

# Plot the distribution of the sample proportions
plt.hist(sample_proportions)
plt.xlabel("Sample Proportion")
plt.ylabel("Frequency")
plt.show()


```

The code above verifies the Central Limit Theorem for the sampling distribution of the sample proportion for a randomly generated population with a known population proportion.As the number of samples increases, the distribution of the sample proportions becomes more normal, even though the population is not normally distributed. This is consistent with the Central Limit Theorem, which states that the distribution of sample proportions will be approximately normal, regardless of the distribution of the population, as long as the sample size is sufficiently large.
