```python
#AYUSH VADALIA  60009220096  D1-1
import numpy as np
def calculate_percentiles(data, percentiles):

    result = {}
    for p in percentiles:
        result[f"{p}th percentile"] = np.percentile(data, p)
    return result
data = [30, 40, 72, 83, 25, 10, 50, 90, 60, 15, 5, 9, 34, 23, 67, 80, 67, 45]
percentiles_to_calculate = [30, 50, 75, 90]
percentile_results = calculate_percentiles(data, percentiles_to_calculate)
for percentile, value in percentile_results.items():
    print(f"{percentile}: {value}")
```

```python
#AYUSH VADALIA  60009220096  D1-1
import pandas as pd
apartments_data = [91, 79, 66, 98, 127, 139, 154, 147, 192, 88, 97, 92, 87, 142, 127, 184, 145, 162, 95, 89, 86, 98, 145, 129, 149, 158, 241]
intervals = [(66, 87), (88, 109), (220, 241)]
frequency_df = pd.DataFrame(index=['Frequency'], columns=[f"{interval[0]}-{interval[1]}" for interval in intervals])
for interval in intervals:
    lower, upper = interval
    frequency_df[f"{lower}-{upper}"] = len([x for x in apartments_data if lower <= x <= upper])
print("Frequency Distribution:")
print(frequency_df)
print()
modal_value = np.argmax(np.bincount(apartments_data))
print(f"Modal Value: {modal_value}")
mean_value = np.mean(apartments_data)
print(f"Mean: {mean_value}")

print("\nComparison:")
print(f"The modal value is {modal_value} and the mean is {mean_value}.")
```

```python
#AYUSH VADALIA  60009220096  D1-1
import numpy as np
heights = np.array([51, 38, 79, 46, 57])
mean_height = np.mean(heights)
sum_squared_diff = np.sum((heights - mean_height) ** 2)
variance = sum_squared_diff / (len(heights) - 1)
std_deviation = np.sqrt(variance)

print(f"Standard Deviation: {std_deviation:.2f} cm")
```

```python
#AYUSH VADALIA  60009220096  D1-1
mileage_data = [3600,4200,4700,4900,5300,5700,6700,7300,7700,8100,8300,8400,8700
                ,8900,9300,9500,9700,10000,10300,10500,10700,
                10800,11000,11300,11500,11800,12100,12700,12900,13100,13500,13800,
                14600,14900,16300,17200,18500,20300]
sorted_mileage = np.sort(mileage_data)
data_range = np.max(sorted_mileage) - np.min(sorted_mileage)

# Calculate interquartile range (IQR)
q1 = np.percentile(sorted_mileage, 25)
q3 = np.percentile(sorted_mileage, 75)
iqr = q3 - q1

# Display results
print(f"Range: {data_range} miles")
print(f"Interquartile Range (IQR): {iqr} miles")
```

```python
#AYUSH VADALIA  60009220096  D1-1
import math


mean_group1 = 32.11
variance_group1 = 68.09
mean_group2 = 19.75
variance_group2 = 71.14


std_dev_group1 = math.sqrt(variance_group1)
std_dev_group2 = math.sqrt(variance_group2)
cv_group1 = (std_dev_group1 / mean_group1) * 100
cv_group2 = (std_dev_group2 / mean_group2) * 100

# Print results
print(f"Group 1 - Coefficient of Variation: {cv_group1:.2f}%")
print(f"Group 2 - Coefficient of Variation: {cv_group2:.2f}%")
if cv_group1 < cv_group2:
    print("Training Program A has less relative variability.")
elif cv_group2 < cv_group1:
    print("Training Program B has less relative variability.")
else:
    print("Both training programs have similar relative variability.")
```

```python
#AYUSH VADALIA  60009220096  D1-1
tomato_weights = [6.3, 7.2, 7.3, 8.1, 7.8, 6.8, 7.5, 7.8, 7.2, 7.5, 8.1, 8.2, 8.0, 7.4, 7.6, 7.7, 7.6, 7.4, 7.5, 8.4, 7.4, 7.6, 6.2, 7.4]

# Calculate average weight
average_weight = sum(tomato_weights) / len(tomato_weights)

# Calculate standard deviation
std_deviation = math.sqrt(sum((weight - average_weight) ** 2 for weight in tomato_weights) / len(tomato_weights))

# Chef's criteria
target_average = 7.5
max_std_deviation = 0.5

# Chef's decision
if abs(average_weight - target_average) <= max_std_deviation and std_deviation < max_std_deviation:
    print("The chef will accept the tomatoes.")
else:
    print("The chef will not accept the tomatoes.")

# Print calculated values
print(f"Average Weight: {average_weight:.2f} ounces")
print(f"Standard Deviation: {std_deviation:.2f} ounces")
```

```python
#AYUSH VADALIA  60009220096  D1-1
import matplotlib.pyplot as plt
day_scores = np.array([99, 56, 78, 55.5, 32, 90, 80, 81, 56, 59, 45, 77, 84.5, 84, 70, 72, 68, 32, 79, 90])

# Test scores for the night class
night_scores = np.array([98, 78, 68, 83, 81, 89, 88, 76, 65, 45, 98, 90, 80, 84.5, 85, 79, 78, 98, 90, 79, 81, 25.5])

# (a) Descriptive statistics for the day class
day_min = np.min(day_scores)
day_max = np.max(day_scores)
day_median = np.median(day_scores)
day_q1 = np.percentile(day_scores, 25)
day_q3 = np.percentile(day_scores, 75)

# (b) Descriptive statistics for the night class
night_min = np.min(night_scores)
night_max = np.max(night_scores)
night_median = np.median(night_scores)
night_q1 = np.percentile(night_scores, 25)
night_q3 = np.percentile(night_scores, 75)

# (c) Percentage of data between different quartiles
def percentage_between(arr, start, end):
    return np.sum((arr >= start) & (arr <= end)) / len(arr) * 100

day_percentage_q1_median = percentage_between(day_scores, day_min, day_q1)
day_percentage_median_q3 = percentage_between(day_scores, day_q1, day_median)
day_percentage_q3_max = percentage_between(day_scores, day_median, day_q3)
day_percentage_max = percentage_between(day_scores, day_q3, day_max)
day_percentage_q1_max = percentage_between(day_scores, day_q1, day_max)

night_percentage_q1_median = percentage_between(night_scores, night_min, night_q1)
night_percentage_median_q3 = percentage_between(night_scores, night_q1, night_median)
night_percentage_q3_max = percentage_between(night_scores, night_median, night_q3)
night_percentage_max = percentage_between(night_scores, night_q3, night_max)
night_percentage_q1_max = percentage_between(night_scores, night_q1, night_max)

# (d) Create box plots
plt.boxplot([day_scores, night_scores], labels=['Day Class', 'Night Class'])
plt.title('Box Plots of Test Scores')
plt.xlabel('Class')
plt.ylabel('Test Scores')
plt.show()

# (e) Compare box plots
day_iqr = day_q3 - day_q1
night_iqr = night_q3 - night_q1

print(f"Interquartile Range (IQR) for Day Class: {day_iqr}")
print(f"Interquartile Range (IQR) for Night Class: {night_iqr}")

# Determine which box plot has the widest spread for the middle 50% of the data
if day_iqr > night_iqr:
    print("Day Class has the widest spread for the middle 50% of the data.")
else:
    print("Night Class has the widest spread for the middle 50% of the data.")
```
