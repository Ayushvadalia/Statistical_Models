#Q1

```python
X=[80,100,120,140,160,180,200,220,240,260]
Y=[70,65,90,95,110,115,120,140,155,150]
```

```python
# prompt: coefficient of correlation between X and y

import numpy as np
x = np.array(X)
y = np.array(Y)
r = np.corrcoef(x, y)[0,1]
print("The coefficient of correlation between X and Y is:", r)
```

```python
import scipy.stats as stats


r, p_value = stats.pearsonr(x, y)

if p_value < 0.05:
    print("The correlation is statistically significant.")
else:
    print("The correlation is not statistically significant.")
```

```python
import statsmodels.api as sm

# Load data
data = {'X':[80,100,120,140,160,180,200,220,240,260],
        'Y':[70,65,90,95,110,115,120,140,155,150]}  # Your data on consumption and income


# Specify the model
model = sm.OLS(data['Y'], sm.add_constant(data['X']))

# Estimate the parameters
results = model.fit()

# Print the results
print(results.summary())

# Access the estimated coefficients
intercept = results.params[0]
slope = results.params[1]

# Calculate the MPC
mpc = slope

print("Estimated MPC:", mpc)
```

```python
r, p_value = stats.pearsonr(x, y)

if p_value < 0.05:
    print("The correlation is statistically significant.")
else:
    print("The correlation is not statistically significant.")
```

```python

r_squared = results.rsquared

print("R-squared:", r_squared)
```
